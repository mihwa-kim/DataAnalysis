```python
import pandas as pd
from sklearn import datasets

iris_data = datasets.load_iris()
```

```python
iris_data
```

```python
df_data = pd.DataFrame(iris_data['data'],columns=iris_data['feature_names'])
df_data.sample(5)
```

```python
df_target = pd.DataFrame(iris_data['target'], columns=['species'])
df_target.sample(5)
```

```python
df_target['species'].unique()
```

```python
df = pd.concat([df_data, df_target], axis=1)
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

df.corr()
```

```python
df[['sepal length (cm)',
  'sepal width (cm)',
  'petal length (cm)',
  'petal width (cm)']].corr()
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(df[['sepal length (cm)',
  'sepal width (cm)',
  'petal length (cm)',
  'petal width (cm)']].corr(), annot=True)
```

```python
corr.values
```

```python
column_names = ['sepal length','sepal width','petal length','petal width']
```

```python
sns.set(font_scale=1)

hm = sns.heatmap(corr.values,
                 cbar=True,
                 annot=True,
                 square=True,
                 fmt='.2f',
                 annot_kws={'size':10},
                 yticklabels=column_names,
                 xticklabels=column_names
            )

plt.tight_layout()
plt.show()
```

```python
df['petal width (cm)'].plot(kind='hist')
plt.title('Iris Petal Width', fontsize=14, y=1.01)
plt.ylabel('Value')
plt.xlabel('Width')
```

```python
df['sepal width (cm)'].plot(kind='hist')
plt.title('Iris Sepal Width', fontsize=14, y=1.01)
plt.ylabel('Value')
plt.xlabel('Width')
```

```python
pie_labels = df['species'].value_counts().index.tolist()
pie_values = df['species'].value_counts().values.tolist()
print(pie_labels)
print(pie_values)
```
